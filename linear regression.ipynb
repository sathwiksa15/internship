```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
weeks_data=[1,2,3,4,5]
sales_data=[2,4,5,4,5]
print(weeks_data,sales_data)
```

```python
df= pd.DataFrame({"weeks":weeks_data,"sales":sales_data})
df
```

```python
X = df["weeks"].values
Y = df["sales"].values
```

```python
mean_x=np.mean(X)
mean_y=np.mean(Y)
n= len(X)
denom=0
numer=0
for i in range(n):
    numer+= (X[i]-mean_x)*(Y[i]-mean_y)
    denom+= (X[i]-mean_x)**2
m=numer/denom
c=mean_y-(m*mean_x)
print(f"m{m}")
print(f"c{c}")
```

```python
max_x= np.max(X)+1
min_x= np.min(X)-1
x= np.linspace(min_x,max_x)
y= (m * x)+c
plt.plot(x, y, color='#8c250f', label='Regression line', linestyle='--')
plt.scatter(X, Y, color='black', label='Data Point')
plt.xlabel("Weeks")
plt.ylabel("Sales")
plt.title("LR Demo")
plt.grid(True,alpha=0.4,linestyle='--')
plt.legend(loc='best')
```
